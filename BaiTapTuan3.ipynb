W3A15

```python
x = float(input())

int_part = int(x)             
frac = x - int_part           

if x >= 0:
    floor_val = int_part
else:
    floor_val = int_part if frac == 0 else int_part - 1

if x >= 0:
    ceil_val = int_part if frac == 0 else int_part + 1
else:
    ceil_val = int_part

if abs(frac) < 0.5:
    round_val = floor_val
else:
    round_val = ceil_val

print(ceil_val, floor_val, round_val)
```
